```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados processados
df = pd.read_csv("../dados_processados/terremotos_processados.csv", parse_dates=["data_hora"])

# Adicionar coluna "mês"
df["mes"] = df["data_hora"].dt.to_period("M").astype(str)

# Agrupar por mês e nível (leve ou forte)
resumo = df.groupby(["mes", "nivel"]).size().unstack(fill_value=0)

# Plotar gráfico de barras empilhadas
plt.figure(figsize=(10, 6))
resumo.plot(kind="bar", stacked=True, colormap="coolwarm")
plt.title("Frequência de Terremotos por Mês (Leves vs Fortes)")
plt.xlabel("Mês")
plt.ylabel("Número de Terremotos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
