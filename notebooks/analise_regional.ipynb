```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../dados_processados/terremotos_processados.csv")

df.head()
```

```python
#Região com maior número de terremotos
mais_frequente = df['local'].value_counts().head(10)
print("Top 10 regiões com mais registros de terremotos:")
print(mais_frequente)
```

```python
#Reigões com terremotos mais graves
graves = df[df['magnitude'] >= 6]
mais_graves = graves['local'].value_counts().head(10)
print("Top 10 regiões com terremotos fortes:")
print(mais_graves)
```

```python
#visualização dos terremotos fortes por região
mais_graves.plot(kind='barh', title='Top 10 Regiões com Terremotos Fortes', color='darkred')
plt.xlabel("Quantidade")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
