```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados processados
df = pd.read_csv("../dados_processados/terremotos_processados.csv", parse_dates=["data_hora"])

# Verificar as colunas e as primeiras linhas
print("Colunas disponíveis:", df.columns)
print(df.head())

# Definir terremotos severos como aqueles com magnitude > 6.0
df_severos = df[df['magnitude'] > 6.0]

# Contar os terremotos severos por local
locais_severos = df_severos['local'].value_counts()

# Plotar gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=locais_severos.index, y=locais_severos.values, palette="viridis")
plt.title("Quantidade de Terremotos Severos por Local")
plt.xlabel("Local")
plt.ylabel("Número de Terremotos Severos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
